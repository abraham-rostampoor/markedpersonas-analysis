```python
import sys
import os
import pprint
from collections import Counter
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '../methods')))
from marked_words import marked_words 
```

```python
df_gpt_4o = pd.read_csv('../new_data/gpt-4o/gpt-4o_personas.csv')
df_gpt_4 = pd.read_csv('../new_data/gpt-4/gpt-4_personas.csv')
df_gpt_4_turbo = pd.read_csv('../new_data/gpt-4-turbo/gpt-4-turbo_personas.csv')
df_gemini = pd.read_csv('../new_data/gemini/gemini_personas.csv')
df_gpt_35_turbo = pd.read_csv('../new_data/gpt-3.5-turbo/gpt-3.5-turbo-0125_personas.csv')
df_llama3 = pd.read_csv('../new_data/llama3/meta-llama/Llama-3-70b-chat-hf_personas.csv')
df_mixtral = pd.read_csv('../new_data/mixtral/mistralai/Mixtral-8x22B-Instruct-v0.1_personas.csv')

dfs = [df_gpt_4o, df_gpt_4, df_gpt_4_turbo, df_gemini, df_gpt_35_turbo, df_llama3, df_mixtral]
```

```python
for df in dfs:
    df['text_clean'] = df['text'].str.lower().str.replace('[^\w\s]', '', regex=True)
    df.dropna(inplace=True)
```

```python
gpt_4o_mw = pd.read_csv('../new_data/mp-analysis-data/llm-mw/gpt4o_mw.csv')
gpt_4_mw = pd.read_csv('../new_data/mp-analysis-data/llm-mw/gpt4_mw.csv')
gpt_4_turbo_mw= pd.read_csv('../new_data/mp-analysis-data/llm-mw/gpt4-turbo_mw.csv')
gemini_mw = pd.read_csv('../new_data/mp-analysis-data/llm-mw/gemini1.5-flash_mw.csv')
gpt_35_turbo_mw = pd.read_csv('../new_data/mp-analysis-data/llm-mw/gpt3.5-turbo_mw.csv')
llama3_mw = pd.read_csv('../new_data/mp-analysis-data/llm-mw/llama-3_mw.csv')
mixtral_mw = pd.read_csv('../new_data/mp-analysis-data/llm-mw/mixtral_mw.csv')

dfs_mw = [gpt_4o_mw, gpt_4_mw, gpt_4_turbo_mw, gemini_mw,
              gpt_35_turbo_mw, llama3_mw, mixtral_mw]
```

```python
models = ['GPT-4o', 'GPT-4', 'GPT-4 Turbo', 'Gemini-1.5 Flash', 'GPT-3.5 Turbo', 'Llama-3', 'Mixtral']
races = ['a Black', 'an Asian', 'a Latino', 'a Middle-Eastern', 'a White']
genders = ['M', 'W', 'N']
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import defaultdict

weighted_scores = defaultdict(dict)

for i, model in enumerate(models):
    df = dfs[i]
    df_mw = dfs_mw[i]
    
    for index, row in df_mw.iterrows():
        race = row['race']
        gender = row['gender']

        if pd.notna(race) and pd.notna(gender) and race in races and gender in genders:
            mw_captured_words = eval(row['mw_captured_words'])
            
            prompts = df[(df['race'] == race) & (df['gender'] == gender)]['text_clean']
            
            total_weighted_score = 0
            
            # print(f"{model}: {race} {gender}\n")
            # print(mw_captured_words)
            for word, z_score in mw_captured_words:
                word_frequency = sum(prompts.str.contains(word))
                weighted_score = word_frequency * z_score
                total_weighted_score += weighted_score
            
            demographic_group = f"{race} {gender}"
            weighted_scores[demographic_group][model] = total_weighted_score

for demographic_group in weighted_scores:
    total_prompts = len(dfs[0])
    for model in models:
        weighted_scores[demographic_group][model] /= total_prompts

weighted_scores_df = pd.DataFrame(weighted_scores)

output_dir = '../new_figures/weighted_scores_figures'
os.makedirs(output_dir, exist_ok=True)

for demographic_group in weighted_scores_df.columns:
    plt.figure(figsize=(10, 6))

    bars = weighted_scores_df[demographic_group].plot(kind='bar', color='orange', width=0.4)
    
    for bar in bars.patches:
        bars.annotate(format(bar.get_height(), '.2f'), 
                      (bar.get_x() + bar.get_width() / 2, bar.get_height()), 
                      ha='center', va='center', 
                      size=10, xytext=(0, 8),
                      textcoords='offset points')
    
    # plt.title(f'Weighted Scores for {demographic_group}')
    plt.xlabel('LLMs')
    plt.ylabel('Weighted Score')
    plt.xticks(rotation=45)
    plt.ylim(0, max(weighted_scores_df[demographic_group])*1.5)
    plt.grid(axis='y', linestyle='--', linewidth=0.7)
    plt.savefig(f'{output_dir}/{demographic_group.replace(" ", "_")}_weighted_score.pdf')
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

intersectional_groups = weighted_scores_df.columns.tolist()

bar_width = 0.2
index = np.arange(len(intersectional_groups))

colors = ['orange', 'brown', 'green', 'pink', 'blue', 'purple', 'grey']

gap = 0.7  

index_with_gaps = index * (1 + gap)

plt.figure(figsize=(16, 8))

for group_idx, demographic_group in enumerate(intersectional_groups):
    
    group_scores = [(weighted_scores_df[demographic_group][i], models[i], colors[i]) for i in range(len(models))]
    group_scores.sort()

    sorted_scores = [x[0] for x in group_scores]
    sorted_models = [x[1] for x in group_scores]
    sorted_colors = [x[2] for x in group_scores]
    
    for i in range(len(sorted_models)):
        bar = plt.bar(index_with_gaps[group_idx] + i * bar_width, sorted_scores[i], bar_width, 
                      label=sorted_models[i] if group_idx == 0 else "", color=sorted_colors[i])
        plt.text(index_with_gaps[group_idx] + i * bar_width, sorted_scores[i] + 0.02, 
                 f'{sorted_scores[i]:.2f}', ha='center', va='bottom', 
                 fontsize=10, rotation=90)

plt.xlabel('Intersectional Demographic Groups', fontsize=16, fontweight='bold')
plt.ylabel('Weighted Score', fontsize=16, fontweight='bold')
plt.ylim(0, max(weighted_scores_df.max()) * 1.1)
# plt.title('LLM Weighted Scores by Intersectional Demographic Groups', fontsize=16, fontweight='bold')
plt.xticks(index_with_gaps + bar_width * len(models) / 2, intersectional_groups, fontsize=12, rotation=45)
plt.yticks(fontsize=14, fontweight='bold')
plt.legend()
plt.grid(axis='y', linestyle='--', linewidth=0.7)

plt.tight_layout()
plt.savefig(f'{output_dir}/all_weighted_score.pdf')
plt.show()
```
