```python
from marked_words import marked_words
import pandas as pd

def pprint(dic):
    for word in sorted(dic,key=lambda x: x[1],reverse=True):
        print("%s, %.2f" % (word[0],word[1]))
```

```python
df1 = pd.read_csv('data/gpt4_main_generations.csv')
nb_df = pd.read_csv('data/gpt4_generations_nb.csv')
nb_df=nb_df.drop(0)
```

```python
nb_df=nb_df.drop(0)
```

```python
df=pd.concat([df1,nb_df])
```

```python
nb_counts = 0
man_c = 0
woman_c = 0
for i,k in df.iterrows():
#     print(k)
    try:
        if 'independent' in k['text']:
            if k['gender']=='M':
                man_c+=1
            if k['gender']=='W':
                woman_c+=1
            if k['gender']=='N':
                nb_counts+=1
    except TypeError:
        continue

#         print(k)
```

```python
print(nb_counts,man_c,woman_c)
```

```python
# Top words for intersectional groups
for race in df['race'].unique():
    print('\n Top words for %s N \n-------' % race)
    pprint(marked_words(df, [race, 'N'], ['race', 'gender'],['a White','M']))
```
