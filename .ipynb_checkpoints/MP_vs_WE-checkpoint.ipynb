```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
from gensim.models import Word2Vec
from scipy.stats import ttest_ind

def calculate_bias_frequency(data, bias_terms):
    data['bias_count'] = data['text'].apply(lambda x: sum([1 for term in bias_terms if term in x.lower()]))
    return data['bias_count'].mean()

def calculate_weat_effect_size(model, target_words_1, target_words_2, attribute_words_1, attribute_words_2):
    def get_vector(word):
        return model.wv[word] if word in model.wv else np.zeros(model.vector_size)
    
    target_vectors_1 = np.array([get_vector(word) for word in target_words_1])
    target_vectors_2 = np.array([get_vector(word) for word in target_words_2])
    attribute_vectors_1 = np.array([get_vector(word) for word in attribute_words_1])
    attribute_vectors_2 = np.array([get_vector(word) for word in attribute_words_2])
    
    s1 = np.mean(cosine_similarity(target_vectors_1, attribute_vectors_1)) - np.mean(cosine_similarity(target_vectors_1, attribute_vectors_2))
    s2 = np.mean(cosine_similarity(target_vectors_2, attribute_vectors_1)) - np.mean(cosine_similarity(target_vectors_2, attribute_vectors_2))
    
    effect_size = (s1 - s2) / np.std(np.concatenate([cosine_similarity(target_vectors_1, attribute_vectors_1),
                                                      cosine_similarity(target_vectors_1, attribute_vectors_2),
                                                      cosine_similarity(target_vectors_2, attribute_vectors_1),
                                                      cosine_similarity(target_vectors_2, attribute_vectors_2)]))
    return effect_size

def analyze_bias_word2vec(data_path):
    df = pd.read_csv(data_path)
    texts = df['text'].tolist()
    sentences = [text.split() for text in texts]
    model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, workers=4)
    
    target_words_1 = ['man', 'male', 'he', 'him']
    target_words_2 = ['woman', 'female', 'she', 'her']
    attribute_words_1 = ['career', 'corporation', 'salary', 'office']
    attribute_words_2 = ['home', 'parents', 'children', 'family']
    
    weat_effect_size = calculate_weat_effect_size(model, target_words_1, target_words_2, attribute_words_1, attribute_words_2)
    return weat_effect_size

bias_terms = ['black', 'white', 'asian', 'latino', 'middle-eastern']

datasets = {
    "DV2": "data/dv2/dv2_main_generations.csv",
    "DV3": "data/dv3/dv3_main_generations.csv",
    "GPT3.5": "data/chatgpt/chatgpt_main_generations.csv",
    "GPT4": "data/chatgpt/chatgpt_main_generations.csv",
    "GPT4o": "data/gpt-4o/gpt4o_all_personas.csv",
    "Llama3": "data/llama3/generated_personas_llama3.csv",
    "Mixtral": "data/mixtral/generated_personas_mixtral.csv"
}

marked_personas_results = {}
word_embeddings_results = {}

for model, path in datasets.items():
    try:
        df = pd.read_csv(path)
        marked_personas_results[model] = calculate_bias_frequency(df, bias_terms)
        word_embeddings_results[model] = analyze_bias_word2vec(path)
    except Exception as e:
        print(f"Error analyzing {model}: {e}")

df_marked_personas = pd.DataFrame.from_dict(marked_personas_results, orient='index', columns=['Bias Frequency'])
df_word_embeddings = pd.DataFrame.from_dict(word_embeddings_results, orient='index', columns=['WEAT Effect Size'])

df_comparison = df_marked_personas.merge(df_word_embeddings, left_index=True, right_index=True)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
df_comparison['Bias Frequency'].plot(kind='bar', color='skyblue')
plt.title('Marked Personas Bias Frequency')
plt.ylabel('Average Bias Count')
plt.xlabel('Model')

plt.subplot(1, 2, 2)
df_comparison['WEAT Effect Size'].plot(kind='bar', color='salmon')
plt.title('Word Embeddings WEAT Effect Size')
plt.ylabel('WEAT Effect Size')
plt.xlabel('Model')

plt.tight_layout()
plt.show()

# Display results
print(df_comparison)

# Statistical comparison
t_stat, p_value = ttest_ind(df_comparison['Bias Frequency'], df_comparison['WEAT Effect Size'])
print(f"T-statistic: {t_stat}, P-value: {p_value}")

# Interpretation
if p_value < 0.05:
    print("There is a statistically significant difference between the two methods.")
else:
    print("There is no statistically significant difference between the two methods.")
```
