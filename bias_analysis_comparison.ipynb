```python
# bias_analysis_comparison.ipynb

import os
import pandas as pd
import subprocess
import matplotlib.pyplot as plt


datasets = {
    'DV2': 'data/dv2/dv2_main_generations.csv',
    'DV3': 'data/dv3/dv3_main_generations.csv',
    'GPT3.5': 'data/chatgpt/chatgpt_main_generations.csv',
    'GPT4': 'data/chatgpt/chatgpt_main_generations.csv',
    'GPT4o': 'data/gpt-4o/gpt4o_all_personas.csv',
    'Llama3': 'data/llama3/generated_personas_llama3.csv',
    'Mixtral': 'data/mixtral/generated_personas_mixtral.csv'
}
def clean_dataset(file_path):
    try:
        df = pd.read_csv(file_path)
  
        if 'text_clean' not in df.columns:
            df['text_clean'] = df['text'].str.lower().str.replace('[^\w\s]','',regex=True)
        df = df.dropna(subset=['text_clean'])
        
  
        df['text_clean'] = df['text_clean'].astype(str)
        
        return df
    except Exception as e:
        print(f"Error cleaning dataset {file_path}: {e}")
        return None


paths = [
    'data/gpt-4o/gpt4o_all_personas.csv',
    'data/llama3/generated_personas_llama3.csv',
    'data/mixtral/generated_personas_mixtral.csv'
]

for path in paths:
    cleaned_df = clean_dataset(path)
    if cleaned_df is not None:
        cleaned_df.to_csv(path, index=False)
        print(f"Cleaned and saved dataset: {path}")
    else:
        print(f"Failed to clean dataset: {path}")

results = []

for name, filepath in datasets.items():
    print(f'Analyzing dataset: {name}')
    try:
        result = subprocess.run(['python', 'bias_analysis_word2vec.py', filepath], capture_output=True, text=True)
        output = result.stdout.strip()
        weat_effect_size = float(output.split(':')[-1])
        results.append((name, weat_effect_size))
    except Exception as e:
        print(f'Error analyzing {name}: {e}')

df_results = pd.DataFrame(results, columns=['Model', 'WEAT Effect Size'])

plt.figure(figsize=(10, 6))
plt.bar(df_results['Model'], df_results['WEAT Effect Size'], color=['#377eb8', '#ff7f00', '#4daf4a', '#f781bf', '#a65628', '#984ea3', '#999999'])
plt.xlabel('Model')
plt.ylabel('WEAT Effect Size')
plt.title('Bias Analysis Comparison Across Models')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('bias_analysis_comparison.png')
plt.show()

df_results
```

# Bias Analysis Summary and Conclusion
## Summary
The bias analysis conducted across various language models (LLMs) provides insights into the presence of biases within the generated personas. The models analyzed include DV2, DV3, GPT3.5, GPT4, GPT4o, Llama3, and Mixtral. Utilized the Word Embedding Association Test (WEAT) effect size to measure biases, with the results visualized in a bar chart and summarized in a table.

### DV2:
- WEAT Effect Size: -0.001897
- The effect size is close to zero, indicating minimal detectable bias in the generated personas.

### DV3:
- WEAT Effect Size: 0.002070
- Similar to DV2, DV3 shows a slight positive bias, but it remains relatively low.

### GPT3.5:
- WEAT Effect Size: 0.004702
- This model shows a slightly higher positive bias compared to DV2 and DV3.

### GPT4:
- WEAT Effect Size: 0.023435
- GPT4 has the highest positive bias among the models analyzed, suggesting a noticeable level of bias in the generated personas.

### GPT4o:
- WEAT Effect Size: 0.001637
- GPT4o exhibits a minimal positive bias, similar to DV2 and DV3.

### Llama3:
- WEAT Effect Size: 0.017718
- Llama3 demonstrates a significant positive bias, though less than GPT4.

### Mixtral:
- WEAT Effect Size: -0.039946
- Mixtral has a notable negative bias, contrasting with the positive biases observed in other models.

## Conclusion
Based on the analysis we identify varying levels of bias across different LLMs. Notably, GPT4 shows the highest positive bias, while Mixtral exhibits a significant negative bias. DV2, DV3, and GPT4o display minimal biases, indicating better performance in generating unbiased personas. Llama3 also shows a notable positive bias but to a lesser extent than GPT4.

The presence of biases in LLMs can impact their applications, particularly in sensitive areas such as social and cultural contexts. Continuous monitoring and mitigation strategies are necessary to ensure fair and unbiased outputs from these models. Future work should focus on refining these models to minimize biases further and explore additional methods to detect and address biases beyond race and gender, including age and other demographic factors.
